```python
from sklearn.svm import SVC 
from sklearn.datasets import make_blobs

import numpy as np
import matplotlib.pyplot as plt


X_blobs2, y_blobs2  = make_blobs(n_samples=1000, random_state=4, center_box=(-2, 2),cluster_std=0.4, centers=2)
```

### Функция для визуализации

```python
#kernels = ("linear", "poly", "rbf") 
def visualize_svm_hyperplane(clf, X, y, frac=0.3):
    
    clf.fit(X, y)
    plt.figure(figsize=(10, 8))
    
    
    plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.Paired)
    
    ax = plt.gca()
    
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    xx = np.linspace(xlim[0], xlim[1], 30)
    yy = np.linspace(ylim[0], ylim[1], 30)
    YY, XX = np.meshgrid(yy, xx)
    xy = np.vstack([XX.ravel(), YY.ravel()]).T
    Z = clf.decision_function(xy).reshape(XX.shape)
    
    ax.contour(
    XX, YY, Z, colors="k", levels=[-1, 0, 1], alpha=0.5, linestyles=["--", "-", "--"]
)
    
    ax.scatter(
    clf.support_vectors_[:, 0],
    clf.support_vectors_[:, 1],
    s=100,
    linewidth=1,
    facecolors="none",
    edgecolors="k",
)
    plt.show()
```

##  Линейное ядро

```python
svc_linear = SVC(kernel="linear")

visualize_svm_hyperplane(svc_linear, X_blobs2, y_blobs2)
```

## Полиномиальное ядро (2 степени)


```python
svc_poly_2 = SVC(kernel="poly", degree=2)
visualize_svm_hyperplane(svc_poly_2, X_blobs2, y_blobs2)
```

## Полиномиальное ядро (3 степени - по умолчанию)

```python
svc_poly_3 = SVC(kernel="poly")

visualize_svm_hyperplane(svc_poly_3, X_blobs2, y_blobs2)
```

## Полиномиальное ядро (5 степень)

```python
svc_poly_5 = SVC(kernel="poly", degree=5)

visualize_svm_hyperplane(svc_poly_5, X_blobs2, y_blobs2)
```

## RBF:

```python
svc_rbf = SVC(kernel="rbf")

visualize_svm_hyperplane(svc_rbf, X_blobs2, y_blobs2)
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
from sklearn.preprocessing import KBinsDiscretizer
from sklearn.model_selection import train_test_split
import time

import os
import pandas as pd

from sklearn.ensemble import RandomForestClassifier
```

## Загрузка данных и подготовка функции дял оценки точности

```python
def compute_metrics(name, classifier, X_train, y_train, y_true=y_test):
    
    start = time.perf_counter()
    classifier.fit(X_train, y_train)
    finish = time.perf_counter()
    
    time_estimated = f"{finish - start:.2f} sec"
    
    y_pred = classifier.predict(X_test)
    
    accuracy = accuracy_score(y_pred=y_pred, y_true=y_true)
    precision = precision_score(y_pred=y_pred, y_true=y_true)
    recall = recall_score(y_pred=y_pred, y_true=y_true)
    f1 = f1_score(y_pred=y_pred, y_true=y_true)
    
    metrics = {
        "Accuracy": [accuracy],
        "Precision": [precision],
        "Recall": [recall],
        "F1 score": [f1],
        "Fitting time": [time_estimated]
    }
    
    metrics_df = pd.DataFrame(metrics)
    metrics_df.index.name = "Value"
    
    
    metrics_df.style.set_table_attributes("style='display:inline'").set_caption(name)
    return metrics_df
```

```python
DATA_PATH = "./hw_Trees_data//"
x_train_path = os.path.join(DATA_PATH, "x_spam_train.csv")
y_train_path = os.path.join(DATA_PATH,  "y_spam_train.csv")
test_path = os.path.join(DATA_PATH, "x_spam_test.csv")

discretizer = KBinsDiscretizer(n_bins=10, strategy="kmeans", encode="ordinal")

X = pd.read_csv(x_train_path, index_col="Id")
x_data = discretizer.fit_transform(X.values)
X_encoded = pd.DataFrame(data=x_data, columns=X.columns)

y = pd.read_csv(y_train_path)["Expected"]

```

## Разбиение

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, train_size=0.9)
```

### Подбор гиперпараметров

```python
from sklearn.model_selection import GridSearchCV

parameters_grid = {"kernel": ["poly", "linear", "rbf"],
                  "degree": [2, 3, 5]}

grid_search = GridSearchCV(SVC(), param_grid=parameters_grid, scoring="f1", cv=3, verbose=1000)

grid_search.fit(X_encoded, y)
best_parameters = grid_search.best_params_
```

Обучение модели с лучшими гиперпараметрами

```python
svc_best = SVC(**best_parameters)


compute_metrics("SVC classifier metrics", svc_best, X_train, y_train)
```

## Обучение RandomForest для сравнения

```python
rf_param_grid = {"n_estimators": [5, 10, 40, 100],
                "max_depth": [2, 5, 10]}

grid_search_rf = GridSearchCV(RandomForestClassifier(), rf_param_grid, cv=3, scoring="f1", verbose=1000)
grid_search_rf.fit(X_encoded, y)

rf_best_parameters = grid_search_rf.best_params_

rfc_best = RandomForestClassifier(**rf_best_parameters)

compute_metrics("Random forest classifier metrics", rfc_best, X_train, y_train)
```

Посмотрим, какие параметры оказались лучшими для SVM:




```python
best_parameters
```

`degree` при `rbf` не учитывается, значит, лушим ядром для SVM оказалась просто Radial Based Function

### Вывод
Время работы Случайного леса меньше на 0.05 секунды, при этом показатели F1 score и Recall выше у SVM, а precision выше у случайного леса не более чем на 0.01. Точность у SVM тоже выше.

Таким образом, на данных из датасета spam модель SVM отработала лучше
