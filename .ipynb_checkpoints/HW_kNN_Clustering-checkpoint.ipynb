Начиная работать с данными, нам необходимо их предобработать и подготовить. В частности, нам необходимо разделить выборку на две: тренировочную и тестовую. Тренировочная выборка необходима для обучения алгоритма, а тестовая для проверки результатов обучения. Обычно используют коэффициент разделения `0.7`.

```python
def train_test_split(X, y:np.ndarray, ratio=0.7, random_state=42):
    # Возвращает X_train, y_train, X_test, y_test
    # X_train и X_test - массив векторов - две части массива X, разделенного в состветсви с коэффициентом ratio
    # y_train и y_test - соответствующие X_train и X_test метки классов
    
    
    # будем пользоваться случайными индексами для разделения датасета
    rng = np.random.RandomState(random_state)
        
    random_indices_permutation = X.index.values
    rng.shuffle(random_indices_permutation)
        
    train_border = int(X.shape[0] * ratio)
    
    train_indices = random_indices_permutation[:train_border]
    test_indices = random_indices_permutation[train_border:]
    
    X_train, y_train = X.loc[train_indices], y[train_indices]
    
    X_test, y_test = X.loc[test_indices], y[test_indices]
    
    assert X_test.shape[0] + X_train.shape[0] == X.shape[0]
    
    return X_train, y_train, X_test, y_test
```

### 2. Метрики (0.5 балла)

Также прежде чем приступать к решению задачи, нам необходимо определиться с метриками, которые позволят нам оценить полученное решение. Для задач классификации мы можем использовать precision, recall и accuracy. Эти метрики считаются для каждого класса. 

__Precision__ отражает то, насколько редко мы ошибаемся, когда говорим, что объект пренадлежит к классу. 

__Recall__ же отражает то, насколько редко классификатор неправильно классифицирует объекты данного класса.

__Accuracy__ отражает то, какую часть выборки классификатор отнес к правильному классу.

```python
def get_precision_recall_accuracy(y_pred, y_true):
    # Возвращает precision, recall и accuracy
    # precision - набор значений метрики precision для каждого класса
    # recall - набор значений метрики recall для каждого класса
    # accuracy - число, отражающее общую точность предсказания
    
    # количество классов необходимо для постройки confusion-матрицы
    # саму матрицу строить не будем, однако вопсользуемся ее содержимым и посчитаем для каждого класса метрики
    num_classes = len(set(y_pred) | set(y_true))
    
    
    precisions_for_these_data = [None for _ in range(num_classes)]
    recalls_for_these_data = [None for _ in range(num_classes)]
    
    # в данном случае классификация бинарная, так что accuracy можно считать по показателям первого класса
    accuracy = []
    
    for class_ in range(num_classes): # классы закодированы в значения [0, количество классов]
        
        TP = np.logical_and((y_pred == class_), (y_true == class_)).sum()

        TN = np.logical_and((y_pred != class_), (y_true != class_)).sum()

        FP = np.logical_and((y_pred == class_), (y_true != class_)).sum()

        FN = np.logical_and((y_pred != class_), (y_true == class_)).sum()
        
        precision = TP / (TP + FP)
        recall = TP / (TP + FN)
        
        precisions_for_these_data[class_] = precision
        recalls_for_these_data[class_] = recall

        accuracy.append(TP + TN / (TP + FP + TN + FN))
    
    return precisions_for_these_data, recalls_for_these_data, accuracy[0]
```

Теперь, имея этот метод, мы можем построить кривые зависимости Precision, Recall и Accuracy от параметра `k`

```python
def plot_precision_recall(X_train, y_train, X_test, y_test, max_k=30):
    ks = list(range(1, max_k + 1))
    
    classes = len(np.unique(list(y_train) + list(y_test)))
    
    precisions = [[] for _ in range(classes)]
    recalls = [[] for _ in range(classes)]
    accuracies = []
    
    for k in ks:
        classifier = KNearest(k)
        classifier.fit(X_train, y_train)
        y_pred = classifier.predict(X_test)
        precision, recall, acc = get_precision_recall_accuracy(y_pred, y_test)
        for c in range(classes):
            precisions[c].append(precision[c])
            recalls[c].append(recall[c])
        accuracies.append(acc)
    
    
    def plot(x, ys, ylabel, legend=True):        
        plt.figure(figsize = (12, 3))
        plt.xlabel("K")
        plt.ylabel(ylabel)
        plt.xlim(x[0], x[-1])
        plt.ylim(np.min(ys)-0.01, np.max(ys)+0.01)
        for cls, cls_y in enumerate(ys):
            plt.plot(x, cls_y, label="Class " + str(cls))
        if legend:
            plt.legend()
        plt.tight_layout()
        plt.show()
    
    plot(ks, recalls, "Recall")
    plot(ks, precisions, "Precision")
    plot(ks, [accuracies], "Accuracy", legend=False)
```

Также для оценки качества классификации построим __ROC-кривую__. Она отражает зависимость __True Positive Rate__ (TPR) от __False Positive Rate__ (FPR) для заранее фиксированного класса. Чем график выше побочной диагонали - тем лучше.

```python
def plot_roc_curve(X_train, y_train, X_test, y_test, max_k=30):
    positive_samples = sum(1 for y in y_test if y == 0)
    ks = list(range(1, max_k + 1))
    curves_tpr = []
    curves_fpr = []
    colors = []
    
    for k in ks:
        colors.append([k / ks[-1], 0, 1 - k / ks[-1]])
        knearest = KNearest(k)
        knearest.fit(X_train, y_train)
        p_pred = [p[0] for p in knearest.predict_proba(X_test)]
        tpr = []
        fpr = []
        for w in np.arange(-0.01, 1.02, 0.01):
            y_pred = [(0 if p > w else 1) for p in p_pred]
            tpr.append(sum(1 for yp, yt in zip(y_pred, y_test) if yp == 0 and yt == 0) / positive_samples)
            fpr.append(sum(1 for yp, yt in zip(y_pred, y_test) if yp == 0 and yt != 0) / (len(y_test) - positive_samples))
        curves_tpr.append(tpr)
        curves_fpr.append(fpr)
    plt.figure(figsize = (7, 7))
    
    for tpr, fpr, c, k in zip(curves_tpr, curves_fpr, colors, ks):
        plt.plot(fpr, tpr, color=c, label=f"k={k}")
    plt.plot([0, 1], [0, 1], linestyle="--")
    plt.xlabel("False positive rate")
    plt.ylabel("True positive rate")
    plt.xlim(-0.01, 1.01)
    plt.ylim(-0.01, 1.01)
    plt.tight_layout()
    plt.legend()
    plt.show()
```

## 3. KNN (2 балла)
Осталось реализовать сам классификатор. Реализуйте его, используя KD-дерево. (При желании можно воспользоваться библиотечной реализацией дерева)

Метод `__init__` принимает на вход количество соседей, по которым предсказывается класс, и размер листьев KD-дерева.

Метод `fit` должен по набору данных и меток "обучать" классификатор. 

Метод `predict_proba` должен предсказывать вероятности классов для заданного набора данных основываясь на классах соседей

```python
from sklearn.neighbors import KDTree
from collections import Counter

class KNearest:
    def __init__(self, n_neighbors=5, leaf_size=30):
        self.leaf_size = leaf_size
        self.n_neighbors = n_neighbors
    
    def fit(self, X, y):
        self.y = y
        self.num_classes = len(np.unique(y))
        
        # Построим дерево для быстрого поиска соседей:
        self.tree = KDTree(X, leaf_size=self.leaf_size)
        
    def predict_proba(self, X):
        # Возвращает матрицу, в которой строки соответствуют элементам X, а столбцы - классам. На пересечении строки и столбца должна быть указана вероятность того, что элемент относится к классу
        # Вероятность рассчитывается как количество ближайших соседей с данным классом деленное на общее количество соседей
        
        # матрица вероятностей (пока пустая)
        proba_matrix = np.empty((X.shape[0], self.num_classes))
        
        
        _, nearest_indices = self.tree.query(X, k=self.n_neighbors) # получим индексы k ближайших соседей для каждого наблюдения
        
        for index in range(X.shape[0]):
            # сосчитаем количество представителей каждого класса среди соседей и сделаем соответствующий массив
            nearest_classes_amount = Counter([self.y[idx] for idx in nearest_indices[index]])
            nearest_classes_inference = np.array([nearest_classes_amount[class_code] for class_code in range(self.num_classes)])
            
            
            proba_matrix[index] = nearest_classes_inference
            
        
        proba_matrix /= self.n_neighbors
        
        return proba_matrix
        
        
    def predict(self, X):
        return np.argmax(self.predict_proba(X), axis=1)
```

### 1. K-means (2.5 балла)

Первый метод, который предлагается реализовать - метод K средних.

__Описание методов__

`fit(X, y=None)` ищет и запоминает в `self.centroids` центроиды кластеров для набора данных.
`predict(X)` для каждого элемента из `X` возвращает номер кластера, к которому относится данный элемент.

__Инициализация кластеров__

Есть несколько вариантов инициализации кластеров. Нужно реализовать их все:
1. `random` - центроиды кластеров являются случайными точками
2. `sample` - центроиды кластеров выбираются случайно из набора данных
3. `k-means++` - центроиды кластеров инициализируются при помощи метода K-means++


```python
class KMeans:
    
    def __init__(self, n_clusters, init="random", max_iter=300, random_state=42):
        self.centroids = None
        self.n_clusters = n_clusters
        self.max_iter = max_iter
        self.init = init
        self.random_state = random_state
    
    @staticmethod
    def _squared_euclidian_distance_kmeans_pp(X, centroid):
        """
        Считает (L_2) ^ 2
        """
        distances = []
        
        for row in X - centroid:
            distances.append(row @ row)
        
        return np.array(distances)
        
    
    def _initialize(self, X, strategy, n_clusters):
        
        rng = np.random.RandomState(self.random_state)
        
        def random_init():
            """
            Выбирает случайных точки, каждая координата генерируется равномерно случайно из размаха
            координате
            
            """
            nonlocal X
            nonlocal rng
            
            max_values_for_features = pd.DataFrame(data=X).max()
            min_values_for_features = pd.DataFrame(data=X).min()
            
            
            random_centroids = np.random.uniform(low=min_values_for_features, 
                                                 high=max_values_for_features, 
                                                 
                                                 size=(self.n_clusters, X.shape[1]))
            return random_centroids
        
        
        def sample_init():
            """
            Генерируется путем слчайного выбора представителей датасета
            """
            
            nonlocal X
            nonlocal rng
            
            
            random_centroids_indices = rng.random_sample.choice(np.arange(len(X.shape[0])), self.n_clusters)
            centroids = X[random_centroids_indices]
            return centroids
        
    
        def pp_init():
            nonlocal X
            nonlocal rng
            
            centroids = [None for _ in range(self.n_clusters)]
            
            # Первая выбирается случайно
            centroids[0] = X[rng.randint(0, X.shape[0])]
            
            # для следующей центроилы считается Евклидово расстояние до нее  от всех точек
            distances_to_clusters = KMeans._squared_euclidian_distance_kmeans_pp(X, centroid=centroids[0])
            
            # расстояние расценивается как вероятность, где чем оно дальше≤ тем дробвь больше
            proportional_probabilities = distances_to_clusters / distances_to_clusters.sum()
            
            
            # выбор на основе данной модели новой точки
            random_index = rng.choice(X.shape[0], 1, p=proportional_probabilities)[0]
            centroids[1] = X[random_index]
            
            for k in range(2, self.n_clusters):
                
                # круг повторяется для n-2 центроид
                # но каждый раз считается только одна колонка для новой центроиды, так как остальные уже подсчитаны
                # здесь уже берутся минимум по каждому расстоянию и также строится неравномерная
                # вероятностная модель
                
                # distances_to_clusters хранит только минимальное расстояние до ближайшей центроиды 
                # (т. к. нет смысла хранить все расстояния, если уже найдено минимальное - его лишь нужно сравнить 
                # с расстоянием до новой центроиды)
                
                # обновляется на каждой итерации
                distances_to_clusters = np.min(np.c_[distances_to_clusters, 
                                                     KMeans._squared_euclidian_distance_kmeans_pp(X, centroid=centroids[k-1])], 
                                               axis=1)
                
                proportional_probabilities = distances_to_clusters / distances_to_clusters.sum()
                random_index = rng.choice(X.shape[0], 1, p=proportional_probabilities)[0]
                
                centroids[k] = X[random_index]
            
            return centroids
        
        INIT = {
            "random": random_init,
            "sample": sample_init,
            "k-means++": pp_init,
        }
        
        
        # конечный автомат по способу инициализации
        self.centroids = INIT[self.init]()
            
        
    
    def fit(self, X, y=None):
        
        # "Expectation stage"
        self._initialize(X, n_clusters=self.n_clusters, strategy=self.init)
        
        steps_remained = self.max_iter
        
        # "Maximization stage"
        while steps_remained > 0:
                        
            labels = self.predict(X)
            
            # инициализация новых центроид
            new_centroids = np.array([X[labels == i].mean(axis=0) for i in range(self.n_clusters)])
            
            # проверка на сходимость
            if np.all(self.centroids == new_centroids):
                break
            
            self.centroids = new_centroids
            steps_remained -= 1
    
    def _get_pairwise_distances(self, X):
        
        L_2_squared_array = [[] for _ in range(self.n_clusters)]
        
        for centroid_index, centroid in enumerate(self.centroids):
            for row in X - centroid:
                L_2_squared_array[centroid_index].append(row @ row)
        
        pairwise_squared_distances = np.array(L_2_squared_array).T
        
        return pairwise_squared_distances
    
    def predict(self, X):
        pairwise_squared_euclidian_distances = self._get_pairwise_distances(X)
        
        return np.argmin(pairwise_squared_euclidian_distances, axis=1)
```

### 2. DBScan (3.5 балла)
В отличии от K-means, DBScan не позволяет задать количество кластеров, на которое будут разбиты данные. Руководствуясь геометрической интерпретацией, он позволяет выделять кластеры более сложной формы.

__Описание методов__

`fit_predict(X, y=None)` для каждого элемента `X` возвращает метку кластера, к которому он относится.

__Возможные метрики__

* `euclidean`
* `manhattan`
* `chebyshev`

__Для быстрого поиска соседей используйте `sklearn.neighbors.KDTree`__

```python
from sklearn.neighbors import KDTree

class DBScan:
    def __init__(self, eps=0.5, min_samples=5, leaf_size=40, metric="euclidean"):
        self.eps = eps
        self.min_samples = min_samples
        self.metric = metric
        self.leaf_size = leaf_size
                
        
    def fit_predict(self, X, y=None):
        self.tree = KDTree(X, leaf_size=self.leaf_size, metric=self.metric)
        
        
        # 0: outlier/unassigned, -2: border, -1: core
        border = -3
        core = -2
        
        labels = [-1 for _ in range(X.shape[0])]
        
        neighbours_per_index = []
        cores = set()
        noncores = set()
        
        for index, point in enumerate(X):
            neighbours_per_index.append(self.tree.query_radius(point.reshape(1, -1), r=self.eps, count_only=False,
                                                              return_distance=False)[0])
            if len(neighbours_per_index[index]) >= self.min_samples:
                cores.add(index)
                labels[index] = core
            else:
                noncores.add(index)
        
        for index in noncores:
            if set(neighbours_per_index[index]) & cores:
                labels[index] = border
        
        
        class_label = 0
        
        for index, point_label in enumerate(labels):
            deq = deque()
            
            if point_label == core:
                labels[index] = class_label
                
                for neighbour_index in neighbours_per_index[index]:
                    
                    if labels[neighbour_index] == core:
                        deq.append(neighbour_index)
                        labels[neighbour_index] = class_label
                        
                    elif labels[neighbour_index] == border:
                        labels[neighbour_index] = class_label
                
                while len(deq) > 0:
                    neighbours = neighbours_per_index[deq.popleft()]
                    
                    for n in neighbours:
                        if labels[n] == core:
                            labels[n] = class_label
                            deq.append(n)
                        
                        elif labels[n] == border:
                            labels[n] = class_label
                class_label += 1
                

        return labels
```

Протестируйте алгоритм на синтетических данных. При необходимости подберите гиперпараметры для достижения лучшего качества кластеризации.

```python
dbscan = DBScan(min_samples=7, eps=.6)
labels_dbscan_X1 = dbscan.fit_predict(X_1)
visualize_clasters(X_1, labels_dbscan_X1)

dbscan = DBScan()
labels_dbscan_X2 = dbscan.fit_predict(X_2)
visualize_clasters(X_2, labels_dbscan_X2)

dbscan = DBScan(eps=0.2)
labels_dbscan_X2_eps = dbscan.fit_predict(X_2)
visualize_clasters(X_2, labels_dbscan_X2_eps)
```

## Сравнения метрик `Davies-Bouldin index` и `Silhouette score`

1. __`Davies-Bouldin index`__
 * показывает меру похожести кластера с его наиболее схожим (отношение внутрикластерного расстояния к внекластерному ==>
 * ==> __чем _меньше_ значение, тем лучше кластеризация__, так как тогда увеличивается разница между кластерами

2. __`Silhouette score`__
* показывает среднюю для каждого наблюдения взвешенную разницу между расстоянием до ближайшего кластера и расстоянием внутри кластера
* ==> __чем _больше_ значение, тем лучше кластеризация__, так как если значение отрицательное, значит, наблюдение приписано заведомо к неоптимальному кластеру, если 0 - в кластерах есть перекрывание, и если стремится к 1 - то расстояние является большим

```python
from sklearn.metrics import davies_bouldin_score
from sklearn.metrics import silhouette_score
```

```python
def get_clustering_score(X, labels, method):
    print(f"Scores for clustering method {method}\n\n\t\t\t\t\t\t\t| [from worse -----> to better direction]:\nDavies-Bouldin index:\t\t\t\t\t| {davies_bouldin_score(X, labels):.3f} [k>0 ----> 0]\nSilhouette Coefficient:\t\t\t\t\t| {silhouette_score(X, labels):.3f} [-1 ---> 1]", end=f"\n\n{'_'*100}\n")
```

_Прим._ - `k>0` означает положительно число

```python
get_clustering_score(X_1, labels_kmeans_X1, "Kmeans (blobs)")
get_clustering_score(X_1, labels_dbscan_X1, "DBSCAN (blobs)")
```

```python
get_clustering_score(X_2, labels_kmeans_X2, "Kmeans (moon-shaped points)")
get_clustering_score(X_2, labels_dbscan_X2_eps, "DBSCAN (moon-shaped points)")
```
